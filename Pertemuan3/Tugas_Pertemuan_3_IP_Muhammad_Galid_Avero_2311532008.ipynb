#1. Import Library

```
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

#2. Persiapkan Gambar

```
img1_path = "kanan.jpeg"
img2_path = "kiri.jpeg"

# Baca sebagai citra warna (BGR)
img1 = cv2.imread(img1_path)
img2 = cv2.imread(img2_path)

print("Ukuran img1:", img1.shape)
print("Ukuran img2:", img2.shape)
```

#3. Menyamakan Ukuran Agar lebih Pasti

```
# Samakan ukuran img2 ke ukuran img1
height, width = img1.shape[:2]
img2_resized = cv2.resize(img2, (width, height))

print("Ukuran img2 sesudah resize:", img2_resized.shape)
```

#4. Menampilkan ke 2 gambar

```
# Konversi BGR ke RGB supaya warna benar di matplotlib
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2_resized, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(img1_rgb)
plt.title("Citra 1")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img2_rgb)
plt.title("Citra 2 (resize)")
plt.axis('off')

plt.tight_layout()
plt.show()
```

#5. Penjumlahan citra dengan OpenCV

```
# Penjumlahan citra: elemen per elemen
img_add = cv2.add(img1, img2_resized)  # otomatis clip ke 255[web:78][web:81][web:85]

# Tampilkan hasil
img_add_rgb = cv2.cvtColor(img_add, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.imshow(img1_rgb)
plt.title("Citra 1")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(img2_rgb)
plt.title("Citra 2")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(img_add_rgb)
plt.title("Hasil Penjumlahan (cv2.add)")
plt.axis('off')

plt.tight_layout()
plt.show()
```

#6. Penjumlahan dengan bobot (blending)

```
alpha = 0.7  # bobot citra 1
beta  = 0.3  # bobot citra 2
gamma = 0    # bias

img_blend = cv2.addWeighted(img1, alpha, img2_resized, beta, gamma)

img_blend_rgb = cv2.cvtColor(img_blend, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(img_add_rgb)
plt.title("Penjumlahan murni (add)")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(img_blend_rgb)
plt.title(f"Blending (alpha={alpha}, beta={beta})")
plt.axis('off')

plt.tight_layout()
plt.show()
```
